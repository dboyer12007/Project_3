```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
final_df = Path("final_df.csv")

# ReadData File and store into Pandas DataFrame
df1 = pd.read_csv(final_df)
df1.head()
```

```python
#Replace the Column Name "Number" with "Rank"...
df1.rename(columns={"Number": "Rank"}, inplace=True)
df1.rename(columns={"country": "Country_of_Release"}, inplace=True)
df1.rename(columns={"label": "Label"}, inplace=True)
df1.rename(columns={"Year": "Released_Year"}, inplace=True)
df1.head()
```

```python
#Delete "genres", "musicbrainz_id", "release_date" column that has no data:
df1.drop(columns=["genres"], inplace=True)
df1.drop(columns=["musicbrainz_id"], inplace=True)
df1.drop(columns=["release_date"], inplace=True)
df1.head()
```

```python
print(df1.dtypes)
```

```python
# File to Load
albumlist_df = Path("albumlist.csv")

# ReadData File and store into Pandas DataFrame
df2 = pd.read_csv(albumlist_df, encoding="latin1") 

df2.head()
```

```python
#Replace the Column Name "Number" with "Rank"...
df2.rename(columns={"Number": "Rank"}, inplace=True)
df2.rename(columns={"Year": "Released_Year"}, inplace=True)
df2.head()
```

```python
print(df2.dtypes)
```

```python
# Combine the data into a single dataset.
combined_data = pd.merge(df1, df2, how="left", on="Album")
combined_data.head()
```

```python
#Eliminate Duplicate Columns
combined_data.drop(columns=["Rank_y"], inplace=True)
combined_data.drop(columns=["Released_Year_y"], inplace=True)
combined_data.drop(columns=["Artist_y"], inplace=True)
combined_data.drop(columns=["Genre_y"], inplace=True)
combined_data.drop(columns=["Subgenre_y"], inplace=True)
combined_data.head()
```

```python
#Replace the Column Name "Rank_x" with "Rank"...
combined_data.rename(columns={"Rank_x": "Rank"}, inplace=True)
combined_data.rename(columns={"Released_Year_x": "Released_Year"}, inplace=True)
combined_data.rename(columns={"Artist_x": "Artist"}, inplace=True)
combined_data.rename(columns={"Genre_x": "Genre"}, inplace=True)
combined_data.rename(columns={"Subgenre_x": "Subgenre"}, inplace=True)
combined_data.head()
```

```python
album_df = combined_data[['Rank', 'Album', 'Released_Year']].copy()
# Add an auto-incrementing Album_ID column
album_df['Album_ID'] = range(1, len(album_df) + 1)
album_df.to_csv("album.csv", index=False)
```

```python
artist_df = combined_data[['Artist']].drop_duplicates().copy()
# Add an auto-incrementing Artist_ID column
artist_df['Artist_ID'] = range(1, len(artist_df) + 1)
artist_df.to_csv("artist.csv", index=False)
```

```python
genre_df = combined_data[['Genre']].drop_duplicates().copy()
# Add an auto-incrementing Genre_ID column
genre_df['Genre_ID'] = range(1, len(genre_df) + 1)
genre_df.to_csv("genre.csv", index=False)
```

```python
subgenre_df = combined_data[['Subgenre']].drop_duplicates().copy()
# Add an auto-incrementing Subgenre_ID column
subgenre_df['Subgenre_ID'] = range(1, len(subgenre_df) + 1)
subgenre_df.to_csv("subgenre.csv", index=False)
```

```python
country_df = combined_data[['Country_of_Release']].drop_duplicates().copy()
# Add an auto-incrementing Country_ID column
country_df['Country_ID'] = range(1, len(country_df) + 1)
country_df.to_csv("country.csv", index=False)
```

```python
label_df = combined_data[['Label']].drop_duplicates().copy()
# Add an auto-incrementing Label_ID column
label_df['Label_ID'] = range(1, len(label_df) + 1)
label_df.to_csv("label.csv", index=False)
```

```python
# Extract the unique Genre data
genre_df = combined_data[['Genre']].drop_duplicates().copy()
# Add a unique ID column
genre_df['Genre_ID'] = range(1, len(genre_df) + 1)

# Save to a CSV file
genre_df.to_csv("album_genre.csv", index=False)
```

```python
subgenre_df = combined_data[['Subgenre']].drop_duplicates().copy()
subgenre_df['Subgenre_ID'] = range(1, len(subgenre_df) + 1)
subgenre_df.to_csv("album_subgenre.csv", index=False)
```

```python
country_df = combined_data[['Country_of_Release']].drop_duplicates().copy()
country_df['Country_ID'] = range(1, len(country_df) + 1)
country_df.to_csv("album_country.csv", index=False)
```

```python
label_df = combined_data[['Label']].drop_duplicates().copy()
label_df['Label_ID'] = range(1, len(label_df) + 1)
label_df.to_csv("album_label.csv", index=False)
```
